# 📊 Exploring Netflix Data (EDA)

## 🎯 Objective
In this project, I explored the Netflix dataset to understand what kind of content is available on the platform. The goal was to identify patterns, trends, and insights about movies and TV shows using data analysis and visualization.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 📁 About the Dataset

This dataset contains information about movies and TV shows available on Netflix.

Some important columns include:
- TYPE → whether it's a Movie or TV Show  
- TITLE → name of the content  
- COUNTRY → where it was produced  
- RELEASE YEAR → year it was released  
- DATE ADDED → when it was added to Netflix  
- RATING → audience rating  

This data helps us understand how Netflix content is distributed.

```python
df = pd.read_csv("/content/netflix_titles.csv")
df.head()
```

## ❓ Questions I Wanted to Explore

Before starting the analysis, I asked a few simple questions:

- Does Netflix have more movies or TV shows?
- Which countries produce the most content?
- How has Netflix content changed over time?
- What kind of ratings are most common?

These questions helped guide my analysis.

```python
df.info()
df.isnull().sum()
```

## 🔍 Data Cleaning

Before analyzing the data, I checked for missing values and made a few adjustments to keep the dataset clean and usable.

```python
df['country'].fillna("Unknown", inplace=True)
df['rating'].fillna("Not Rated", inplace=True)
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows Distribution")
plt.show()
```

### 📊 Observation

From the graph, it’s clear that Netflix has more **movies** than TV shows. This shows that the platform focuses more on movie content.

```python
top_countries = df['country'].value_counts().head(10)

top_countries.plot(kind='bar')
plt.title("Top 10 Countries by Content")
plt.xticks(rotation=45)
plt.show()
```

### 📊 Observation

Some countries clearly produce more content than others. A few countries dominate Netflix’s library.

```python
df['year_added'] = df['date_added'].dt.year

sns.countplot(x='year_added', data=df)
plt.title("Content Added Over Years")
plt.xticks(rotation=45)
plt.show()
```

### 📊 Observation

The number of shows and movies added to Netflix has increased over the years. This reflects the platform’s growth and expansion.



```python
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title("Rating Distribution")
plt.show()
```

### 📊 Observation

Certain ratings appear more frequently, which suggests the type of audience Netflix is targeting.

## 🔍 Key Insights

- Netflix has significantly more movies than TV shows  
- Content production is concentrated in a few countries  
- The platform has grown rapidly in recent years  
- Audience ratings show clear viewing preferences
